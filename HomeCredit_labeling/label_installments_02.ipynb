```python
import pandas as pd
import numpy as np
```

```python
data=r'D:\期末專題相關\HomeCredit\labeling\installments_payments_weighting.csv'
dataf = pd.read_csv(data)
dataf
```

```python
dataf['late_count_2'] = (dataf['late_count']>0)*1
dataf
```

```python
test_sec = dataf['late_count']*dataf['weighting']/dataf['INSTALMENT_count']
dataf['weighting_2'] = np.ceil(test_sec*3/np.max(test_sec))
dataf
```

```python
np.max(dataf['weighting_2'])
```

```python
dataf2 = dataf.groupby('SK_ID_CURR',as_index = False).aggregate({
    'SK_ID_PREV':'count',
    'late_count_2':'sum',
    'weighting_2':'sum'
})
dataf2
```

```python
dataf2.columns = ['SK_ID_CURR','installments_count','installments_Late_count','installments_weight']
dataf2
```

```python
data=r'D:\期末專題相關\HomeCredit\labeling\installments_payments_label.csv'
dataf2.to_csv(data,index = False)
```
