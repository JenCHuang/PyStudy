```python
import pandas as pd
import numpy as np
```

```python
data=r'D:\期末專題相關\HomeCredit\origin\credit_card_balance.csv'
dataf = pd.read_csv(data)
dataf
```

```python
dataf = dataf.drop(columns=['AMT_DRAWINGS_ATM_CURRENT','AMT_DRAWINGS_OTHER_CURRENT','AMT_DRAWINGS_POS_CURRENT',
                            'AMT_INST_MIN_REGULARITY','AMT_PAYMENT_CURRENT',
                            'CNT_DRAWINGS_ATM_CURRENT','CNT_DRAWINGS_CURRENT','CNT_DRAWINGS_OTHER_CURRENT',
                            'CNT_DRAWINGS_POS_CURRENT','CNT_INSTALMENT_MATURE_CUM','NAME_CONTRACT_STATUS',
                            'SK_DPD_DEF'])
dataf.isnull().sum()
```

```python
dataf.describe()
```

```python
dataf['ind_payment'] = dataf['AMT_BALANCE'] - dataf['AMT_TOTAL_RECEIVABLE']
dataf['AMT_CREDIT_LIMIT_ACTUAL'][dataf['AMT_CREDIT_LIMIT_ACTUAL']==0] = np.nan
dataf['ind_p_DRAWINGS'] = dataf['AMT_DRAWINGS_CURRENT'] / dataf['AMT_CREDIT_LIMIT_ACTUAL']
dataf
```

```python
dataf.isnull().sum()
```

```python
dataf = dataf.fillna(0)
dataf['simple_late'] = ((dataf['ind_payment'] + dataf['SK_DPD'])>0)*1
dataf
```

```python
dataf2 = dataf.groupby('SK_ID_PREV',as_index = False).aggregate({
    'SK_ID_CURR':'max',
    'MONTHS_BALANCE':'count',
    'simple_late':'sum',
    'SK_DPD':'max',
    'ind_p_DRAWINGS':'mean'
})
dataf2
```

```python
test_sec = dataf2['simple_late']/dataf2['MONTHS_BALANCE']
test_sec = test_sec - np.mean(test_sec)
dataf2['credit_care_late'] = (test_sec>0)*1
print(sum(dataf2['credit_care_late']))
```

```python
test_sec = (dataf2['simple_late']/dataf2['MONTHS_BALANCE'])*4 + dataf2['ind_p_DRAWINGS']
test_sec = test_sec*10/np.max(test_sec)
dataf2['credit_care_coef'] = np.ceil(test_sec)/10 + dataf2['credit_care_late']
dataf2
```

```python
dataf3 = dataf2.groupby('SK_ID_CURR',as_index = False).aggregate({
    'SK_ID_PREV':'count',
    'credit_care_late':'sum',
    'credit_care_coef':'sum'
})
dataf3.columns = ['SK_ID_CURR','credit_card_count','credit_card_late','credit_card_coef']
dataf3
```

```python
data=r'D:\期末專題相關\HomeCredit\labeling\credit_card_label.csv'
dataf3.to_csv(data,index = False)
```
