```python
import pandas as pd
import numpy as np
```

```python
data=r'D:\期末專題相關\HomeCredit\origin\installments_payments.csv'
dataf = pd.read_csv(data)
dataf
```

```python
dataf.isnull().sum()
```

```python
test_sec = dataf['DAYS_ENTRY_PAYMENT'] - dataf['DAYS_INSTALMENT']
test_sec = (test_sec>0)*test_sec
dataf['entry_late'] = test_sec/np.max(test_sec)
dataf
```

```python
test_sec = dataf['AMT_INSTALMENT'] - dataf['AMT_PAYMENT']
test_sec = (test_sec>0)*test_sec
dataf['PAYMENT_nonpay'] = test_sec/np.max(test_sec)
dataf
```

```python
dataf['late_count']= ((dataf['entry_late']+dataf['PAYMENT_nonpay'])>0)*1
dataf = dataf.fillna(0)
dataf
```

```python
test_sec = dataf['late_count']+1.5*(dataf['entry_late']+dataf['PAYMENT_nonpay'])
print(np.max(test_sec))
print(np.mean(test_sec))
#dataf['weighting'] 
```

```python
dataf['weighting']=np.ceil(test_sec*3/np.max(test_sec))
dataf
```

```python
dataf = dataf.groupby('SK_ID_PREV',as_index = False).aggregate({
    'SK_ID_CURR':'max',
    'NUM_INSTALMENT_NUMBER':'count',
    'late_count':'sum',
    'weighting':'sum'
})
dataf
```

```python
dataf.columns = ['SK_ID_PREV','SK_ID_CURR','INSTALMENT_count','late_count','weighting']
dataf
```

```python
data=r'D:\期末專題相關\HomeCredit\labeling\installments_payments_weighting.csv'
dataf.to_csv(data,index = False)
```
