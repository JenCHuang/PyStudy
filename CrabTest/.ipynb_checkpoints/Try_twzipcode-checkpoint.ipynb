```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import re
```

```python
s = Service("./chromedriver.exe")  
driver = webdriver.Chrome(service=s)
driver.get("https://twzipcode.com/") 
time.sleep(1)

addtext = '桃園市平鎮區新德街14號'
#addtext = '桃園市中壢區龍岡路三段262巷60號'

search_box = driver.find_element(By.ID, "Search_C_T")
search_box.send_keys(addtext)
time.sleep(1)
button = driver.find_element(By.ID, "submit")
button.click()
time.sleep(2)

target_html = str(driver.page_source)

soup = BeautifulSoup(target_html, 'html.parser')
itemlist = soup.find_all("td",{'class':'item'})

collist = ['縣市','鄉鎮','村里']
ctrln=0

for i, k in enumerate(itemlist):
    itemname = str(k).replace('<td class="item">','').replace('</td>','')
    if itemname in collist:
        ctrln=1
        value = str(soup).split(str(k)+'<td>')[1].split("</td></tr>")[0]
        if value[0:2] == '<a' :
            value = value.split('>')[1].split('<')[0]
        print(f'{itemname}: {value}')

if ctrln == 0:
    print('沒有村里資訊')
    
driver.close()
```
