```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
import time
```

```python
s = Service("./chromedriver.exe")  
driver = webdriver.Chrome(service=s)
driver.get("https://www.tgos.tw/tgos/Web/Address/TGOS_Address.aspx") 
time.sleep(1)

addtext = '桃園市平鎮區復旦路二段122號'

search_box = driver.find_element(By.ID, "TGOS_AddressBox_QueryAddress_Value")
search_box.send_keys(addtext)
time.sleep(2)
button = driver.find_element(By.ID, "TGOS_AddressBox_QueryAddress_Execute")
button.click()
time.sleep(2)

driver.switch_to.window(driver.window_handles[1])
time.sleep(2)

page_info = driver.find_element(By.ID, "MapBox")
#print(page_info)
time.sleep(1)
page_text = page_info.text

#target_html = str(driver.page_source)
driver.quit()

for i in range(3):
    print(page_text.split('\n')[i])
```
