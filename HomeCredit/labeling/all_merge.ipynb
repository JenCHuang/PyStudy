```python
test_sec = dataf['bureau_Late_count']/dataf['bureau_count']
test0 = test_sec.fillna(0)
test0
```

```python
test_sec = dataf['POS_CASH_late_count']/dataf['POS_CASH_count']
test1 = test_sec.fillna(0)
test_sec = dataf['installments_Late_count']/dataf['installments_count']
test2 = test_sec.fillna(0)
test_sec = dataf['credit_card_late']/dataf['credit_card_count']
test3 = test_sec.fillna(0)
```

```python
test_sec = test0 + test1 + test2 + test3
dataf['simple_label'] = (test_sec>0)*1
dataf['sum_avg_late'] = test_sec
```

```python
test_sec = dataf['bureau_score']/dataf['bureau_count']
test0 = test_sec.fillna(0)
test_sec = dataf['POS_CASH_weight']/dataf['POS_CASH_count']
test1 = test_sec.fillna(0)
test_sec = dataf['installments_weight']/dataf['installments_count']
test2 = test_sec.fillna(0)
test_sec = dataf['credit_card_coef']/dataf['credit_card_count']
test3 = test_sec.fillna(0)
```

```python
dataf = dataf.fillna(0)
dataf['total_score'] = dataf['bureau_score'] + dataf['POS_CASH_weight'] + dataf['installments_weight'] + dataf['credit_card_coef']
test_sec = test0 + test1 + test2 + test3
dataf['sum_avg_score'] = test_sec
```

```python
dataf2 = dataf[['SK_ID_CURR','simple_label','sum_avg_late','total_score','sum_avg_score']]
dataf2
```
