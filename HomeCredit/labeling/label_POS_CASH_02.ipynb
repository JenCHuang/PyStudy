```python
import pandas as pd
import numpy as np
```

```python
data=r'D:\期末專題相關\HomeCredit\labeling\POS_CASH_balance_weighting.csv'
dataf = pd.read_csv(data)
dataf
```

```python
np.max(dataf['weighting'])
```

```python
dataf['late_label'] = ((dataf['test_INSTALMENT']+dataf['test_SK_DPD'])>0)*1
dataf
```

```python
np.max(dataf['late_label'])
```

```python
dataf['late_1y_or_last'] = ((dataf['in_one_year_late']+dataf['last'])>0)*1
dataf
```

```python
data_t = dataf.groupby('SK_ID_CURR',as_index = False).aggregate({'SK_ID_PREV':'count',
                                                                 'late_label':'sum',
                                                                 'late_1y_or_last':'sum',
                                                                 'weighting':'sum'})
data_t
```

```python
data_t.columns = ['SK_ID_CURR','SK_count','Late_count','total_weighting']
data_t
```

```python
data=r'D:\期末專題相關\HomeCredit\labeling\POS_CASH_balance_label.csv'
data_t.to_csv(data,index = False)
```
