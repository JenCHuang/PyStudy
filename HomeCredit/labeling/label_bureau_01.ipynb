```python
data = "gdrive/MyDrive/BDSE_Report_Info/HomeCredit_firstView/data/bureau.csv"
dataf = pd.read_csv(data)
dataf
```

```python
print(dataf.info())
print("------------")
print(dataf.isnull().sum())
```

```python
print(dataf['SK_ID_BUREAU'].drop_duplicates().count())
sum(np.array(dataf['DAYS_CREDIT'] > -365))
```

```python
dataf=dataf[['SK_ID_CURR','CREDIT_ACTIVE','DAYS_CREDIT','CREDIT_DAY_OVERDUE','AMT_CREDIT_MAX_OVERDUE','CNT_CREDIT_PROLONG','AMT_CREDIT_SUM_OVERDUE','DAYS_CREDIT_UPDATE']]
dataf = dataf.replace(np.nan,0)
dataf['in_one_year'] = (dataf['DAYS_CREDIT'] > -365)*1
dataf
```

```python
## CREDIT_ACTIVE 信用局记录在案的信贷的当前状态
## DAYS_CREDIT 客户申请信用局信贷距离当前申请天数
## CREDIT_DAY_OVERDUE 根据客户申请贷款的时间点来说，信用局信贷违约天数
## AMT_CREDIT_MAX_OVERDUE 迄今为止，信用局信贷最高过期额度
## CNT_CREDIT_PROLONG 信用局信贷延长了多少次
## AMT_CREDIT_SUM_OVERDUE 信用局信贷当前逾期金额
## DAYS_CREDIT_UPDAT 最后一次来源于信用局的信息距离当今申请贷款天数
dataf.groupby('SK_ID_CURR')['CREDIT_ACTIVE'].count()
```

```python
print(dataf['CREDIT_ACTIVE'].drop_duplicates())
print(sum(dataf['CREDIT_ACTIVE']=='Bad debt'))
```

```python
dataf['bad_debt']= (dataf['CREDIT_ACTIVE']=='Bad debt')*1
print(sum(dataf['bad_debt']))
dataf = dataf.drop(columns=['CREDIT_ACTIVE'])
dataf
```

```python
test_sec = dataf['CREDIT_DAY_OVERDUE']/np.mean(dataf['CREDIT_DAY_OVERDUE'])
dataf['coef_DAY_OVERDUE'] = test_sec/np.max(test_sec)
dataf
```

```python
test_sec = dataf['AMT_CREDIT_SUM_OVERDUE']/np.mean(dataf['AMT_CREDIT_SUM_OVERDUE'])
dataf['coef_SUM_OVERDUE'] = test_sec/np.max(test_sec)
dataf
```

```python
test_sec = dataf['AMT_CREDIT_MAX_OVERDUE']/np.mean(dataf['AMT_CREDIT_MAX_OVERDUE'])
dataf['coefadding_mix'] = (test_sec/np.max(test_sec) + dataf['CNT_CREDIT_PROLONG']/np.max(dataf['CNT_CREDIT_PROLONG']))*0.5
dataf
```

```python
dataf['bad_and_late'] = ((dataf['coef_DAY_OVERDUE'] + dataf['coef_SUM_OVERDUE'] + dataf['bad_debt']) > 0)*1
dataf
```

```python
dataf = dataf[['SK_ID_CURR','in_one_year','bad_debt','coef_DAY_OVERDUE','coef_SUM_OVERDUE','coefadding_mix','bad_and_late']]
dataf
```

```python
data = "gdrive/MyDrive/bureau_weighting.csv"
dataf.to_csv(data)
```
