```python
dataf = dataf[['SK_ID_PREV','SK_ID_CURR','AMT_RECIVABLE','AMT_RECEIVABLE_PRINCIPAL',
               'AMT_PAYMENT_CURRENT','AMT_INST_MIN_REGULARITY','SK_DPD']]
dataf = dataf.fillna(0)
dataf
```

```python
test1 = (dataf['AMT_RECIVABLE']<dataf['AMT_RECEIVABLE_PRINCIPAL'])*1
test2 = (dataf['AMT_PAYMENT_CURRENT']<dataf['AMT_INST_MIN_REGULARITY'])*1
test3 = (dataf['SK_DPD']>0)
dataf['late_count']=(test1+test2+test3 > 0)*1
dataf
```

```python
data_t2 = dataf.groupby('SK_ID_PREV',as_index = False).aggregate({'SK_ID_CURR':'max',
                                                                  'AMT_RECIVABLE':'count',
                                                                  'late_count':'sum'})
data_t2
```

```python
data_t2['late_label'] = (dataf['late_count']>0)*1
data_t2
```

```python
data_t2 = data_t2.groupby('SK_ID_CURR',as_index = False).aggregate({
    'SK_ID_PREV':'count',
    'AMT_RECIVABLE':'sum',
    'late_count':'sum',
    'late_label':'sum'
})
data_t2
```

```python
data_t2.columns = ['SK_ID_CURR','Loan_count','INSTALMENT_count','late_count_total','late_count_inLoan']
data_t2
```

```python
data=r'C:\Users\Student\My_iSpan\ziSpanProject_data\processed\processed_credit_card_balance.csv'
data_t2.to_csv(data,index = False)
```
