```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data = "gdrive/MyDrive/Home_Credit_TEST/bureau_weighting.csv"
dataf = pd.read_csv(data)
dataf
```

```python
dataf = dataf.drop(columns=['Unnamed: 0'])
dataf
```

```python
dataf['bureau_simple'] = ((dataf['coef_DAY_OVERDUE'] + dataf['coef_SUM_OVERDUE'] + dataf['bad_and_late']) > 0 )*1
dataf
```

```python
score = (dataf['coef_DAY_OVERDUE'] + dataf['coef_SUM_OVERDUE'] + dataf['bureau_simple']) * (1 + dataf['in_one_year']) + dataf['coefadding_mix']
print(np.max(score))
print(np.mean(score))
```

```python
dataf['bureau_score'] = np.ceil(score*6/np.max(score))
dataf
```

```python
print(np.max(dataf['bureau_score']))
print(np.mean(dataf['bureau_score']))
```

```python
dataf2 = dataf.groupby('SK_ID_CURR',as_index = False).aggregate({'bad_and_late':'count',
                                  'bureau_simple':'sum',
                                  'bureau_score':'sum'})
dataf2
```

```python
np.max(dataf2['bureau_simple'])
```

```python
dataf2.columns = ['SK_ID_CURR','bureau_count','bureau_Late_count','bureau_score']
dataf2
```

```python
data="gdrive/MyDrive/bureau_label.csv"
dataf2.to_csv(data,index = False)
```
