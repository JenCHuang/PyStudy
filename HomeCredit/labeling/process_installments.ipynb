```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=r'C:\Users\Student\My_iSpan\ziSpanProject_data\origin\installments_payments.csv'
dataf = pd.read_csv(data)
dataf
```

```python
dataf.isnull().sum()
```

```python
test_sec = dataf['DAYS_ENTRY_PAYMENT'] - dataf['DAYS_INSTALMENT']
dataf['entry_late'] = (test_sec>0)*test_sec
dataf
```

```python
col = 'entry_late'
print(np.max(dataf[col]))
print(sum(dataf[col]>0))
dataf[col].hist()
```

```python
test_sec = dataf['AMT_INSTALMENT'] - dataf['AMT_PAYMENT']
dataf['PAYMENT_nonpay'] = (test_sec>0)*test_sec
dataf
```

```python
col = 'PAYMENT_nonpay'
print(np.max(dataf[col]))
print(sum(dataf[col]>0))
dataf[col].hist()
```

```python
dataf = dataf.fillna(0)
dataf['late_count']= ((dataf['entry_late']+dataf['PAYMENT_nonpay'])>0)*1
dataf
```

```python
print(sum(dataf['late_count']))
```

```python
dataf = dataf.groupby('SK_ID_PREV',as_index = False).aggregate({
    'SK_ID_CURR':'max',
    'NUM_INSTALMENT_NUMBER':'count',
    'late_count':'sum'
})
dataf
```

```python
dataf.columns = ['SK_ID_PREV','SK_ID_CURR','INSTALMENT_count','late_count_total']
dataf['late_label'] = (dataf['late_count_total']>0)*1
dataf
```

```python
dataf = dataf.groupby('SK_ID_CURR',as_index = False).aggregate({
    'SK_ID_PREV':'count',
    'INSTALMENT_count':'sum',
    'late_count_total':'sum',
    'late_label':'sum'
})
dataf
```

```python
dataf.columns = ['SK_ID_CURR','Loan_count','INSTALMENT_count','late_count_total','late_count_inLoan']
dataf
```

```python
data=r'C:\Users\Student\My_iSpan\ziSpanProject_data\processed\processed_installments_payments.csv'
dataf.to_csv(data,index = False)
```
