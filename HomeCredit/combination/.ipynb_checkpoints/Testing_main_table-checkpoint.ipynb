<h2> Adding One feature </h2>

```python
data = r'C:\Users\Student\My_iSpan\ziSpanProject_data\origin\application_train.csv'
datafa=pd.read_csv(data)
datafa
```

```python
col_n = [i for i in range(96,116)]
col_n = [0] + col_n
print(col_n)
```

```python
datafa = datafa.iloc[:,col_n]
datafa['papers'] = datafa.iloc[:,1:].sum(axis=1)
datafa
```

```python
datafa=datafa[['SK_ID_CURR','papers']]
datafa['paper%']=datafa['papers']/20
datafa
```

```python
data = r'C:\Users\Student\My_iSpan\ziSpanProject_data\origin\application_test.csv'
datafb=pd.read_csv(data)
datafb
```

```python
col_n = [i for i in range(95,115)]
col_n = [0] + col_n
print(col_n)
```

```python
datafb = datafb.iloc[:,col_n]
datafb['papers'] = datafb.iloc[:,1:].sum(axis=1)
datafb
```

```python
datafb=datafb[['SK_ID_CURR','papers']]
datafb['paper%']=datafb['papers']/20
datafb
```

```python
datafc = pd.concat([datafa,datafb])
datafc = datafc[['SK_ID_CURR','paper%']]
datafc
```

```python
dataf = pd.merge(dataf,datafc,how='outer')
dataf
```

```python
dataf.isnull().sum()
```
